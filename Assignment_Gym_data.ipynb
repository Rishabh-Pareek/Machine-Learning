### gym Data

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv('google_review_ratings.csv', index_col='User')
```

```python
data.head()
```

```python
data.info()
```

```python
# Converting feature-Category 11 object to float
data['Category 11'] = pd.to_numeric(data['Category 11'],errors='coerce')
```

```python
# As we can see data.info there are only 2 values in unnamed:25 so we will remover this feature 
data.drop(columns=['Unnamed: 25'], inplace=True)
data.head()
```

* Correlation between features

```python
plt.subplots(figsize=(29,15))
sns.heatmap(data.corr(),annot = True)
```

```python
data.describe()
```

* Here we don't have a target column. we create one by taking avg rating of cols in a row.

```python
sns.distplot(data['Category 1'],kde = True)
```

```python
df = pd.DataFrame(data.mean(axis=1),columns=['avg'])

data['membership'] = df['avg'].apply(lambda x: 0 if x <= 2 else 1)
```

```python
data['membership'].value_counts()
```

* if avg <= 2 : 0 will not take membership
* elif avg > 2: 1 will take membership

### Spliting Data

```python
feature_data = data.drop(columns = ['membership'],axis=1)
target_data = data.membership
```

```python
from sklearn.model_selection import train_test_split
```

```python
trainX,testX,trainY,testY = train_test_split(feature_data,target_data)
```

### Logistic Regression

```python
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
```

```python
pipeline = make_pipeline(SimpleImputer(),StandardScaler(),LogisticRegression())
```

```python
pipeline.fit(trainX,trainY)
```

```python
pipeline.score(testX,testY)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
y_pred = pipeline.predict(testX)
```

```python
confusion_matrix(testY,y_pred)
```

### RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.feature_selection import SelectKBest, f_classif
```

```python
pipeline = make_pipeline(SimpleImputer(),StandardScaler(),SelectKBest(k=10, score_func=f_classif), RandomForestClassifier(n_estimators=100))
```

```python
pipeline.fit(trainX,trainY)
```

```python
pipeline.score(testX,testY)
```

```python
y_pred = pipeline.predict(testX)
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(testY,y_pred)
```

* Hyper-parameter tuning GridSearchCV

```python
from sklearn.model_selection import GridSearchCV
```

```python
params = {'selectkbest__k':[10,15,24],'randomforestclassifier__n_estimators':[100,200,300]}
```

```python
gs = GridSearchCV(pipeline,param_grid = params,cv = 5,n_jobs = 4)
```

```python
gs.fit(trainX,trainY)
```

```python
gs.best_params_
```

```python
gs.best_score_
```
